# Feature Engineering
Our goal: 
To predict overallRating
We want to create different sets of features that might help the model make accurate predictions.
By separating features into groups, we can compare which set works best when training the model.

```python
# ===============================
# 1. Simple Feature Engineering
# ===============================

# Create main_genre
imdb['main_genre'] = imdb['genres'].str.split(",").str[0]

# ===============================
# 2. Split train and test
# ===============================

from sklearn.model_selection import train_test_split
X = imdb[['numVotes', 'runtimeMinutes', 'startYear', 'main_genre']]
Y = imdb['overallRating']

X_train, X_test, y_train, y_test = train_test_split(
    X, Y, test_size=0.2, random_state=100, shuffle=True
)

# ===============================
# 3. Label Encode main_genre
# ===============================

from sklearn.preprocessing import LabelEncoder
le = LabelEncoder()

X_train['main_genre_le'] = le.fit_transform(X_train['main_genre'])

# ===============================
# 4. Feature Sets
# ===============================

X_train_base = X_train[['main_genre_le']]
X_train_movie = X_train[['numVotes', 'runtimeMinutes', 'startYear']]
X_train_full = X_train[['numVotes', 'runtimeMinutes', 'startYear', 'main_genre_le']]

X_train_set = {
    'base': X_train_base,
    'movie': X_train_movie,
    'full': X_train_full,
}
```

# 6. Cross Validation

We use regression because movie ratings are numerical values.
In this project, we compare several regression models to predict the IMDb rating:

- Linear Regression

- Ridge Regression

- Lasso Regression

```python
# ===============================
# 6. Cross Validation
# ===============================

# We use regression because overallRating is numerical (continuous value).
# We will consider a few models:
# * Linear Regression
# * Ridge Regression
# * Lasso Regression

import sklearn.linear_model as lm
from sklearn.pipeline import Pipeline
from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import cross_val_score

models = {
    'Linear Regression': lm.LinearRegression(),
    'Ridge Regression alpha=1': lm.Ridge(alpha=1),
    'Ridge Regression alpha=0.1': lm.Ridge(alpha=0.1),
    'Lasso Regression alpha=1': lm.Lasso(alpha=1),
    'Lasso Regression alpha=0.1': lm.Lasso(alpha=0.1),
}

cv_results = []

for model_name, model in models.items():
    for X_name, X_temp in X_train_set.items():

        pipeline = Pipeline([
            ('scaler', StandardScaler()),
            ('regressor', model)
        ])

        cv_scores = cross_val_score(
            estimator = pipeline,
            X = X_temp,
            y = y_train,
            cv = 5,
            scoring = 'neg_mean_squared_error'
        )

        mean_cv_mse = -cv_scores.mean()   # convert negative MSE back to positive MSE

        cv_results.append({
            'Feature Set': X_name,
            'Model': model_name,
            'Mean CV MSE': mean_cv_mse,
        })

cv_results_df = pd.DataFrame(cv_results)
print(cv_results_df)
```

# 7. Fit the best model and check the error test set

```python
# ===============================
# 7. Fit the Best Model and Check Error on Test Set
# ===============================

from sklearn.preprocessing import StandardScaler
from sklearn.metrics import mean_squared_error
import sklearn.linear_model as lm
import numpy as np

# -------------------------------------------------------
# 1. Choose the best model (example: Ridge Regression α=0.1)
# -------------------------------------------------------
best_model = lm.Ridge(alpha=0.1)

# -------------------------------------------------------
# 2. Final feature set (using main_genre, not main_genre_le)
# -------------------------------------------------------
X_train_final = X_train[['numVotes', 'runtimeMinutes', 'startYear', 'main_genre']]
X_test_final  = X_test[['numVotes', 'runtimeMinutes', 'startYear', 'main_genre']]

# -------------------------------------------------------
# 3. Convert categorical columns to dummies
# -------------------------------------------------------
X_train_final = pd.get_dummies(X_train_final)
X_test_final  = pd.get_dummies(X_test_final)

# IMPORTANT: align columns so train/test match
X_train_final, X_test_final = X_train_final.align(X_test_final, join='left', axis=1, fill_value=0)

# -------------------------------------------------------
# 4. Standardize the data
# -------------------------------------------------------
scaler = StandardScaler()
X_train_scaled = scaler.fit_transform(X_train_final)
X_test_scaled  = scaler.transform(X_test_final)

# -------------------------------------------------------
# 5. Fit the model on training data
# -------------------------------------------------------
best_model.fit(X_train_scaled, y_train)

# -------------------------------------------------------
# 6. Predict on test data + compute MSE
# -------------------------------------------------------
y_pred = best_model.predict(X_test_scaled)
mse = mean_squared_error(y_test, y_pred)

print(f"MSE: {mse}")
# Compute RMSE
rmse = np.sqrt(mse)
print("RMSE:", rmse)
```

# 8. Visualize Predictions vs Actual Ratings

```python
# ===============================
# 8. Predictions vs Actual (Scatter Plot)
# ===============================

plot_df = pd.DataFrame({
    'Actual': y_test,
    'Predicted': y_pred
})

(
    ggplot(plot_df, aes(x='Actual', y='Predicted')) +
    geom_point(alpha=0.4) +
    geom_abline(color='red') +
    labs(title="Predicted vs Actual Ratings",
         x="Actual Rating",
         y="Predicted Rating")
)
```

# 10. Error Distribution(Check if the model is biased)

```python
# ===============================
# 10. Error Distribution
# ===============================

errors = y_test - y_pred

(
    ggplot(pd.DataFrame({'error': errors}), aes(x='error')) +
    geom_histogram(bins=20, fill='blue', alpha=0.5) +
    labs(title="Distribution of Prediction Errors",
         x="Error (Actual - Predicted)",
         y="Count")
)
```
